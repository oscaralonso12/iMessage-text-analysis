```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('white')
import time, datetime
from collections import Counter
```

# Yamz

```python
df = pd.read_csv("iMessage.csv")
df.head()
```

```python
#Identify senders
df.loc[(df['is_from_me'] == 1, "is_from_me")] = "Oscar"
df.loc[(df['is_from_me'] == 0, "is_from_me")] = "Yamz"

#Get rid of null cells
df = df.dropna(subset = ['text'])
df = df[df['text'] !=0]
```

```python
#Create a column for the number of words in messages
df['words'] = df['text'].apply(lambda x: len(str(x).split(" ")))

#Create another column for the number of characters in messages
df['characters'] = df['text'].str.len()

#Lowercase for all words
df['text'] = df['text'].apply(lambda x: " ".join(x.lower() for x in x.split()))
```

```python
#Group dataframe to see most common messages sent by each sender
groupsMessage = df.set_index('is_from_me')
groupsMessage.sort_index(inplace=True)

#Most common messages G sent
mostCommonG = groupsMessage.loc["Oscar", "text"].value_counts().head(23)
mostCommonG = pd.DataFrame(mostCommonG)

#Most common messages L sent
mostCommonL = groupsMessage.loc["Yamz", "text"].value_counts().head(23)
mostCommonL = pd.DataFrame(mostCommonL)

display("Most common words used by Oscar", mostCommonG) 
print("")
display("Most common words used by Yamz", mostCommonL)
```
